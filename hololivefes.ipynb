```python

import json
import threading

# Define a semaphore with the desired number of permits
semaphore = threading.Semaphore(3)

# Define a function that represents the thread's task
def update_ticket_stock(id, quantity):

    # Acquire the semaphore permit
    semaphore.acquire()

    """
    [
    {
        "stage_id": 1,
        "stage_name": "Stage 1 Day 1",
        "ticket_price": 7950,
        "ticket_stock": 1500
    },
    {
        "stage_id": 2,
        "stage_name": "Stage 2 Day 1",
        "ticket_price": 7950,
        "ticket_stock": 1500
    },
    {
        "stage_id": 3,
        "stage_name": "Stage 3 Day 2",
        "ticket_price": 7950,
        "ticket_stock": 1500
    },
    {
        "stage_id": 4,
        "stage_name": "Honeyworks Stage Day 2",
        "ticket_price": 7950,
        "ticket_stock": 1500
    }
    ]

    """

    try:
        # Access the file here
        with open("db.json", "r+") as file:
            # read the file
            data = file.read()
            # convert the data to a dictionary
            data = json.loads(data)
            
            # find the stage with the given id
            for stage in data:
                if stage["stage_id"] == id:
                    # update the ticket stock
                    stage["ticket_stock"] -= quantity
                    break

            # write the updated data back to the file
            file.seek(0)
            file.write(json.dumps(data, indent=4))
            file.truncate()
        
    finally:
        # Release the semaphore permit
        semaphore.release()
```

```python
# Create multiple instances of the thread
threads = []
for i in range(100):
    thread = threading.Thread(target=update_ticket_stock, args=(1, 1))
    threads.append(thread)

# Start all the threads
for thread in threads:
    thread.start()
```
